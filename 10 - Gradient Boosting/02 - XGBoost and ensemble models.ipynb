# Weak learners

**Weak learners** are basic models that do not perform so well by themselves, either because they have a high bias (low degree of freedom models, for example) or because they have too much variance to be robust (high degree of freedom models, for example). The idea of ensemble methods is to try reducing bias and/or variance of such weak learners by combining several of them together in order to create a strong learner (or ensemble model) that achieves better performances.

In the following exemple, we will use a "decision stumps." A decision stump is simply a decision tree where the whole tree is just one node.

```python
import numpy as np

from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
from sklearn.tree import DecisionTreeClassifier 

class Stump:
    def __init__(self):
        self.gtLabel = None
        self.ltLabel = None
        self.splitThreshold = None
        self.splitFeature = None

    def predict(self, listpoint):
        return np.array([self.predict_single(x) for x in listpoint])
        
    def predict_single(self, point):
        if point[self.splitFeature] >= self.splitThreshold:
            return self.gtLabel
        else:
            return self.ltLabel

    def __call__(self, point):
        return self.predict(point)

    def errorFunction(self, data,index, t):
        left_side = sum([1.0 for x,y in data if ... != y]) / len(data)
        right_side = 1. - left_side
        return min(left_side, right_side)


    def makeThreshold(self,x, t,index):
        return  1 if x[index] <= t else 0

    def bestThreshold(self,data, index):
        '''Compute best threshold for a given feature. Returns (threshold, error)'''

        thresholds = [x[index] for x,y in data]
        errors = [(threshold, self.errorFunction(data, index, threshold)) for threshold in thresholds]
        return min(errors, key=lambda p: p[1])


    def majorityVote(self,labels):
        try:
            return max(set(labels), key=labels.count)
        except:
            return 0

    def fit(self,X, Y):
        data = list(zip(X,Y))
        # find the index of the best feature to split on, and the best threshold for that index.

        bestThresholds = [(i,) + self.bestThreshold(data, i) for i in range(len(X[0]))]
        feature, thresh, _ = min(bestThresholds, key = lambda p: p[2])

        self.splitFeature = feature
        self.splitThreshold = thresh
        self.gtLabel = self.majorityVote([y for x,y in data if x[self.splitFeature] >= self.splitThreshold])
        self.ltLabel = self.majorityVote([y for x,y in data if x[self.splitFeature] < self.splitThreshold])


```

```python
breast = datasets.load_breast_cancer()
#print(breast.DESCR)

X_train, X_test, y_train, y_test = train_test_split(breast.data, breast.target, test_size=0.2, random_state=42)

stump = Stump()
stump.fit(X_train,y_train)
print(accuracy_score(y_train, stump.predict(X_train)))
print(accuracy_score(y_test, stump.predict(X_test)))
```

```python
# you can use this code as a comparison point
tree = DecisionTreeClassifier(max_depth=1, random_state=0)
tree.fit(X_train,y_train)
print(accuracy_score(y_test, tree.predict(X_test)))


n_nodes = tree.tree_.node_count
children_left = tree.tree_.children_left
children_right = tree.tree_.children_right
feature = tree.tree_.feature
threshold = tree.tree_.threshold

node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
is_leaves = np.zeros(shape=n_nodes, dtype=bool)
stack = [(0, -1)]  # seed is the root node id and its parent depth
while len(stack) > 0:
    node_id, parent_depth = stack.pop()
    node_depth[node_id] = parent_depth + 1

    # If we have a test node
    if (children_left[node_id] != children_right[node_id]):
        stack.append((children_left[node_id], parent_depth + 1))
        stack.append((children_right[node_id], parent_depth + 1))
    else:
        is_leaves[node_id] = True

print("The binary tree structure has %s nodes and has "
      "the following tree structure:"
      % n_nodes)
for i in range(n_nodes):
    if is_leaves[i]:
        print("%snode=%s leaf node." % (node_depth[i] * "\t", i))
    else:
        print("%snode=%s test node: go to node %s if X[:, %s] <= %s else to "
              "node %s."
              % (node_depth[i] * "\t",
                 i,
                 children_left[i],
                 feature[i],
                 threshold[i],
                 children_right[i],
                 ))
```



# Boosting vs Bagging vs stacking
The main hypothesis is that when weak models are correctly combined we can obtain more accurate and/or robust models.

* **Bagging** considers homogeneous weak learners, learns them independently from each other in parallel and combines them following some kind of deterministic averaging process
* **Boosting** considers homogeneous weak learners, learns them sequentially in a very adaptative way (a base model depends on the previous ones) and combines them following a deterministic strategy
* **Stacking** considers heterogeneous weak learners, learns them in parallel and combines them by training a meta-model to output a prediction based on the different weak models predictions

## Focus on Boosting
The intuition behind boosting is that each new model focus its efforts on the most difficult observations to fit up to now, so that we obtain, at the end of the process, a strong learner with lower bias. Boosting is a gerneral technic that is used with success by the the [top solutions](https://www.kaggle.com/c/porto-seguro-safe-driver-prediction/discussion/44629#250927) in Kaggle competition - it is a crucial competence to master by going beyond using black-box library.

Being mainly focused at reducing bias, the base models that are often considered for boosting are models with low variance but high bias. For instance, if we want to use trees as our base models, we will choose most of the time shallow decision trees with only a few depths.

There are two emblematic Boosting algorithms : **Adaboost** and **Gradient Boosting**. Since Gradient Boosting is the algorithm used in XGboost, we will try to understand how it works.

Here are the steps of a simpler version of Gradient Boosting :
* 1 - Fit a simple  decision tree on data
* 2 - Calculate error residuals. Actual target value, minus predicted target value
* 3 - Fit a new model on error residuals as target variable with same input variables
* 4 - Add the predicted residuals to the previous predictions
* 5 - Fit another model on residuals that is still left. Repeat steps 2 to 5 until overfitting or the sum of residuals becomes constant

As you can see, we have the *boosting* logic of having Decision Trees one after the other, while each single Decision Tree is very shallow.

We will now focus on some parameters specific to trees. 

* n_estimators: number of trees you want to build.
* max_depth: determines how deeply each tree is allowed to grow during any boosting round.
* subsample: percentage of samples used per tree.
* gamma: minimum loss reduction required to make a further partition on a leaf node of the tree.

    n_estimators:
        Low value (e.g., 10 or less): This can lead to underfitting. With too few trees, the model may not capture complex patterns in the data, resulting in insufficient predictive power.
        High value (e.g., 100 or more): This can lead to overfitting. When you have too many trees, the model may start memorizing the training data, which can lead to poor generalization on unseen data.

    max_depth:
        Low value (e.g., 2 or 3): This can lead to underfitting. A shallow tree may not be able to capture intricate relationships in the data, resulting in a lack of model complexity.
        High value (e.g., 10 or more): This can lead to overfitting. Deeper trees are more likely to fit noise in the training data, making the model less generalizable.

    subsample:
        Low value (e.g., 0.1 or 0.2): This can lead to overfitting. Using a very small percentage of samples for each tree can result in overemphasizing the impact of individual data points, which can lead to overfitting.
        High value (e.g., 0.8 or 1.0): This can lead to underfitting. With a high subsample percentage, the model may not have enough diversity in the training data for each tree, potentially missing important patterns.

    gamma:
        Low value (e.g., 0): This can lead to overfitting. A low value of gamma means that the model does not require a significant loss reduction to create additional splits in the tree, which can lead to overfitting by making the tree too deep.
        High value (e.g., 1 or more): This can lead to underfitting. A high gamma value requires a substantial loss reduction to create new splits, potentially resulting in overly simple and underfit models.

```python
import numpy as np

def ground_truth(x):
    """Ground truth -- function to approximate"""
    return x * np.sin(x) + np.sin(2 * x)

def gen_data(n_samples=200):
    """generate training and testing data"""
    np.random.seed(13)
    x = np.random.uniform(0, 10, size=n_samples)
    x.sort()
    y = ground_truth(x) + 0.75 * np.random.normal(size=n_samples)
    train_mask = np.random.randint(0, 2, size=n_samples).astype(np.bool_)
    x_train, y_train = x[train_mask, np.newaxis], y[train_mask]
    x_test, y_test = x[~train_mask, np.newaxis], y[~train_mask]
    return x_train, x_test, y_train, y_test

X_train, X_test, y_train, y_test = gen_data(200)

# plot ground truth
x_plot = np.linspace(0, 10, 500)
def plot_data(figsize=(8, 5)):
    fig = plt.figure(figsize=figsize)
    gt = plt.plot(x_plot, ground_truth(x_plot), alpha=0.4, label='ground truth')

    # plot training and testing data
    plt.scatter(X_train, y_train, s=10, alpha=0.4)
    plt.scatter(X_test, y_test, s=10, alpha=0.4, color='red')
    plt.xlim((0, 10))
    plt.ylabel('y')
    plt.xlabel('x')
    
plot_data(figsize=(8, 5))
```

```python
from itertools import islice
from xgboost import XGBRegressor

est = XGBRegressor(max_depth=1,subsample=1,gamma=0.8,n_estimators=1000)
est.fit(X_train, y_train)

ax = plt.gca()

# step over prediction as we added 20 more trees.
plot_data()
plt.plot(x_plot, est.predict(x_plot[:, np.newaxis]), color='r', alpha=0.8)
```

combines the power of gradient boosting with decision trees. The model behind XGBoost is a specific implementation of gradient boosting, which uses decision trees as base learners. The core concept of XGBoost is to build an ensemble of decision trees sequentially, where each tree corrects the errors of its predecessor. Here are some key features of the XGBoost model:

    Gradient Boosting: XGBoost uses a gradient descent optimization technique to minimize a loss function. It builds a series of decision trees, and at each step, the model identifies the gradient of the loss with respect to the predictions made by the current ensemble. New trees are constructed to reduce the loss by moving in the direction of the negative gradient.

    Decision Trees as Base Learners: XGBoost employs shallow decision trees (often referred to as "weak learners") as the base model. These decision trees are typically very simple, with a limited depth, which makes them computationally efficient.

    Regularization: XGBoost incorporates L1 (Lasso) and L2 (Ridge) regularization terms into the objective function to prevent overfitting. Regularization helps control the complexity of the individual trees in the ensemble.

    Parallel and Distributed Computing: XGBoost is designed for parallel and distributed computing, making it efficient for large datasets. It can leverage multiple CPU cores and distributed computing frameworks for scalability.

    Handling Missing Values: XGBoost has built-in support for handling missing values in the dataset, simplifying data preprocessing.

    Custom Objective Functions: XGBoost allows you to define custom objective functions, which makes it flexible and adaptable to various machine learning tasks.

    Pruning: XGBoost uses a "prune" step to remove splits in the tree that do not lead to a significant improvement in the loss function. This helps control the depth of trees and avoids overfitting.

b) Random Forest and boosted trees (such as XGBoost) are both ensemble learning methods, but they have some fundamental differences:

    Base Learners:
        Random Forest: Random Forest builds an ensemble of decision trees, where each tree is constructed independently without knowledge of the other trees. Trees in a Random Forest are typically deep and highly diversified.
        Boosted Trees (e.g., XGBoost): Boosted Trees build an ensemble of decision trees sequentially. Each tree is trained to correct the errors made by the previous trees. These trees are often shallow.

    Independence:
        Random Forest: Trees in a Random Forest are constructed independently, and they are not aware of the errors or predictions of other trees. The final prediction is made by averaging (for regression) or voting (for classification) among the trees.
        Boosted Trees: Trees in a boosted ensemble are trained in a sequential manner, and each tree is dependent on the previous ones. They are designed to address the weaknesses of previous trees and reduce the overall error.

    Bias-Variance Trade-off:
        Random Forest: Random Forest aims to reduce overfitting by creating diverse trees and averaging their predictions. It tends to have lower bias but can still exhibit some variance.
        Boosted Trees: Boosted Trees focus on reducing both bias and variance. They can achieve very low bias and, with proper regularization, maintain low variance, leading to high predictive accuracy.

    Robustness:
        Random Forest: Random Forest is relatively robust to overfitting and noise in the data due to the aggregation of multiple diverse trees.
        Boosted Trees: Boosted Trees can be sensitive to noisy data, as they adapt to the errors in the training set, potentially leading to overfitting.
